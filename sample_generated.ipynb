```python
import pandas as pd

# Load the data
data = pd.read_csv('Student.csv')

# Calculate the average math_score
average_math_score = data['math_score'].mean()

# Print the result
print(f"The average math score is: {average_math_score:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('Student.csv')

# Create the column chart
data.groupby('gender')['reading_score'].mean().plot(kind='bar', color=['blue', 'orange'])

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Average Reading Score')
plt.title('Distribution of Reading Score by Gender')

# Save the plot
plt.savefig('output/output_plot.png')

# Show the plot
plt.show()
```
